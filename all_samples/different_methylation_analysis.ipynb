#### SNP genotype (AA, AB, BB)

```python
snp_dir = "/home/galaxy/project/m6AQTL/2019_10_10/transcriptome_snp/SNPiR/shell/all_samples/results/MatrixEQTL_result"
snp_file = os.path.join(snp_dir, "final_SNP_reorder.csv")
```

```python
df_snp = pd.read_csv(snp_file, index_col=0)
df_snp.head()
```

```python
col_list = df_snp.columns
sample_dict = {}

for index, values in df_snp.iterrows():
    count = Counter(values)
    if (count[1] >1) and (count[2] >1):
        ref_homo_list = [i for i in range(len(values)) if values[i] == 0]
        alt_homo_list = [i for i in range(len(values)) if values[i] == 2]
        sample_dict[index] = [ref_homo_list, alt_homo_list]
```

```python
sample_dict["chr1:725401"]
```

```python
matrix_dir = "/home/galaxy/project/m6AQTL/2019_10_10/transcriptome_snp/SNPiR/shell/all_samples/results/quantification"
matrix_file = os.path.join(matrix_dir, "bedtools_quantification.matrix")
```

```python
df_m6a = pd.read_table(matrix_file)
df_m6a.head()
```

```python
result_dir = "/home/galaxy/project/m6AQTL/2019_10_10/transcriptome_snp/SNPiR/shell/all_samples/results/diff_methylation"
with open("%s/snp_m6a.data" % result_dir, 'w') as fw:
    for snp, index_array in sample_dict.items():
        ref_list, alt_list = index_array[0], index_array[1]
        if snp in snp_m6a_dict:
            m6a_id = snp_m6a_dict[snp]
            m6a_values = df_m6a.loc[m6a_id, :]
            ref_values, alt_values = [str(m6a_values[i]) for i in ref_list], [str(m6a_values[i]) for i in alt_list]
            fw.write("%s\t%s\t%s\t%s\n" % (snp, m6a_id, ",".join(ref_values), ",".join(alt_values)))
```
