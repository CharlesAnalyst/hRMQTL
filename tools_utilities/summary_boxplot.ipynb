```python
import os
import glob
import pandas as pd
import numpy as np
from multiprocessing import Pool
```

```python
def describe_boxplot(df_series):
    min_value = df_series.min()
    max_value = df_series.max()
    df_res = df_series.quantile([0.25, 0.5, 0.75])
    print("N\t%d" % len(df_series))
    print("min\t%f" % min_value)
    print("max\t%f" % max_value)
    print(df_res)
```

```python
def figure_2c():
    x = "/home/xialj/data6/xlj/Differential_m6A/Ratio/DMR/Binji/Filter_FPKM0/Expression/Data_for_plot/data_forplot"
    df = pd.read_table(x, sep="\t")
#     print(df.head())
    organ_list = list(set(df["Organ"].tolist()))
    for organ in organ_list:
        print("#### %s #####" % organ)
        df_sub = df[df.Organ == organ]
        df_c = df_sub[df_sub.Type == "Constitutive m(6)As"]
        df_d = df_sub[df_sub.Type == "Differential m(6)As"]
        print("^^^ Constitutive m(6)As ^^^")
        describe_boxplot(df_c["value"])
        print("^^^ Differential m(6)As ^^^")
        describe_boxplot(df_d["value"])
        print()
figure_2c()
```

```python
def figure_4a():
    x = "/home/xialj/data6/xlj/Gene_Expression_divergence/gene_methylation_level/Rank_m6A_number/Result.txt"
    df = pd.read_table(x, sep="\t", skiprows=2)
    term_list = list(set(df["type"].tolist()))
    for term in term_list:
        print("#### %s #####" % term)
        df_sub = df[df.type == term]
        describe_boxplot(df_sub["cv"])
figure_4a()
```
