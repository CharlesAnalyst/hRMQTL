```python
import os
import sys
import csv
import datetime
import shutil
import random
import glob
import pandas as pd
```

#### 1. stat m6a QTL

##### 1.1 m6a eQTL

```python
file_dir = "/home/galaxy/project/QTL_analysis/SNP_gain_or_loss/pQTL/m6ASNP_result"
file_list = glob.glob("%s/*.txt" % file_dir)
for x in file_list:
    tissue = os.path.basename(x).split(".")[0]
    names = ["SampleID", "UCSCID", "Gene symbol", "Chromosome", "Position", "Strand", "Reference sequence", "Mutation sequence", "Reference score", "Mutation score", "Significance", "Mutation event", "Related mutations", "unknown"]
    df = pd.read_table(x, sep="\t", header=None, skiprows=1, names=names)
    df_gain = df[df["Mutation event"] == "Functional Gain"]
    df_loss = df[df["Mutation event"] == "Functional Loss"]
    print("%s\t%d\t%d\t%d" % (tissue, len(df_gain), len(df_loss), len(df_gain)+len(df_loss)))
```
