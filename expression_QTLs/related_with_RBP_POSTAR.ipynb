```python
def get_dict():
    rbp_list = glob.glob("%s/*.txt" % rbp_dir)
    rbp_dict = {}
    for rbp in rbp_list:
        rbp_name = os.path.basename(rbp).split("_")[0]
        rbp_dict[rbp_name] = rbp_dict.get(rbp_name, []) + [rbp]
    return rbp_dict

def process_one_targetGene(rbp):
    result_file = os.path.join(tmp_dir, os.path.basename(rbp))
    target = os.path.basename(rbp).split(".txt")[0].split("_")[1]
    df = pd.read_table(rbp, sep="\t", header=None)
    df.columns = ["score1", "score2", "Gene ID", "Gene Name", "Gene Type", "Genomic context", "Position", "CLIP-seq technology and peak calling method", "db", ".", "Strand", "Tissue type", "Transcript ID"]
    selected_col = ["Gene ID", "Tissue type", "db", "Position"]
    df = df[selected_col].drop_duplicates()
    df["name"] = df["Gene ID"] + ";" + df["db"] + ";" + df["Position"]
    df["Chr name"] = df["Position"].str.split(":").str[0]
    df["Start"] = df["Position"].str.split(":").str[1].str.split("-").str[0]
    df["End"] = df["Position"].str.split("-").str[1]
    df_result = df[["Gene ID", "Tissue type", "Chr name", "Start", "End", "name"]]
    df_result.to_csv(result_file, sep="\t", header=False, index=False)
    
def process_one_rbp(rbp_name, rbp_list):
    print(rbp_name)
    result_total_file = os.path.join(result_dir, "%s.txt" % rbp_name)
    if (not os.path.exists(result_total_file)) and len(glob.glob("%s/*" % tmp_dir)) == 0:
#     if len(glob.glob("%s/*" % tmp_dir)) == 0:
        pool = Pool(processes=30)
        for rbp in rbp_list:
            pool.apply_async(process_one_targetGene, (rbp, ))
        pool.close()
        pool.join()
        os.system("cat %s/*.txt > %s" % (tmp_dir, result_total_file))
        os.system("rm %s/*" % tmp_dir)
    else:
        print("tmp dir exits file!")
        
def main_method():
    rbp_dict = get_dict()
    for rbp_name, rbp_list in rbp_dict.items():
        if len(rbp_list) > 0:
            process_one_rbp(rbp_name, rbp_list)
        else:
            print(rbp_name)
    os.system("rm -rf %s" % tmp_dir)
#################
main_method()
#################
```

```python
rbp_list = glob.glob("%s/*.bed" % rbp_dir)
eQTL_list = glob.glob("%s/*.bed" % eQTL_dir)
HapMap_NUMBER = 81039058
```

```python
for eqtl in eQTL_list:
    res_list, pool = [], Pool(processes=30)
    for rbp in rbp_list:
#         res_list = fisher_test(eqtl, rbp)
        res = pool.apply_async(fisher_test, (eqtl, rbp, ))
        res_list.append(res)
    pool.close()
    pool.join()
    write_to_file(res_list, os.path.join(result_dir, os.path.basename(eqtl).split("-")[0]))
```

```python
def write_to_file(res_list, res_file):
    res_list = ["%s\n" % x.get() for x in res_list]
    with open(res_file, "w") as fw:
        fw.writelines(res_list)
```

```python
def fisher_test(eqtl_bed, RPB_bed):
         #              RPB_bed       x
        # QTL_SNP         a       m6a - a
        # HapMap_SNP      c       con - c
    QTL_num = stat_file_line_numbers(eqtl_bed)
    a = os.popen("bedtools intersect -a %s -b %s -wa |sort| uniq | wc -l" % (eqtl_bed, RPB_bed)).read()
    c = os.popen("bedtools intersect -a %s -b %s -wa |sort| uniq | wc -l" % (HapMap_snp, RPB_bed)).read()
    b, d = QTL_num - int(a), HapMap_NUMBER - int(c)
    oddsratio, pvalue = stats.fisher_exact([[int(a), b], [int(c), d]])
    line="%s\t%s\t%f\t%f"%(os.path.basename(eqtl_bed).split("-")[0], os.path.basename(RPB_bed), oddsratio, pvalue)
    return line
```

```python
def stat_file_line_numbers(in_file):
    raw_return = os.popen("wc -l %s" % in_file)
    line_num = int(raw_return.read().split()[0])
    return line_num
```
